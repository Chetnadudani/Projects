```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset (change filename if needed)
df = pd.read_csv("all_data.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[df['Order Date'] != 'Order Date']
```

```python
df = df.dropna()
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']

print("Total Sales:", round(df['Sales'].sum(), 2))
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%y %H:%M')
```

```python
top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1])

city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)

city_sales
```

```python
city_sales.plot(kind='bar')
plt.title("Sales by City")
plt.show()
```

```python
Business Insights:

1. The analysis reveals that the MacBook Pro Laptop generated the highest revenue among all products.

2. Sales show a noticeable increase towards the end of the year, suggesting significant holiday-season demand.

3. San Francisco and Los Angeles emerge as the top-performing cities.

4. Premium electronic products account for the majority of total sales.

5. A small number of products contribute a large share of total revenue.
```
